## Exploratory Data Analysis on Boston Housing Data

```python
# Creating environment
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading data
df = pd.read_csv('BostonHousing.csv')
```

```python
# Checking for Synopsis of the data
df.head()
```

```python
# Checking for data structure and types
df.info()
```

```python
# Checking for summary statistics
df.describe()
```

```python
# Visualizaiton of Relationship between features and target variables
# Bin 'zn' and 'age' into categories
df['zn_category'] = pd.cut(df['zn'], bins=[0, 25, 75, 100], labels=['Low', 'Medium', 'High'])
df['age_category'] = pd.cut(df['age'], bins=[0, 40, 70, 100], labels=['Young', 'Middle-aged', 'Old'])
# Ploting figure
plt.figure(figsize = (12,8))
sns.scatterplot(x = df['rm'], y = df['medv'], hue = df['zn'], size = df['age'])
plt.title('Impact of Various factors on House Price in Boston')
plt.xlabel('Avg. Num. of Rooms per Dwelling')
plt.ylabel('Median Value of Owner-occupied Homes in $1000s')
plt.show()
```

```python
# Identifying Missing Values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cmap='viridis', cbar=False, yticklabels=False)
plt.title("Missing Values in Dataset")
plt.show()
```

```python
# Identifying outliers by help of Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['crim', 'rm', 'tax', 'lstat', 'medv']])
plt.xticks(rotation=45)
plt.title("Boxplot for Outlier Detection")
plt.show()
```
